```python
from PIL import Image
import numpy as np
import imagecodecs
from skimage import filters, feature
from scipy.stats import entropy, skew, kurtosis

from pandarallel import pandarallel
pandarallel.initialize()


def calculate_image_features(jp2_bytes, thresholds=range(1, 255, 10)):
    """
    Calculate various image features, including dark pixel percentages, brightness, contrast, texture, edges,
    sharpness, and histogram metrics from a JP2 image.

    Parameters:
    jp2_bytes (bytes): The JP2 image data as bytes.
    thresholds (iterable): Grayscale intensity thresholds for calculating dark pixel percentages.

    Returns:
    dict: Dictionary containing various calculated image metrics.
    """
    try:
        # Decode JP2 image and convert to grayscale
        image_data = imagecodecs.jpeg2k_decode(jp2_bytes)
        image = Image.fromarray(image_data).convert('L')
        pixel_data = np.array(image)
        
        # Initialize dictionary to store all metrics
        metrics = {}

        # 1. Dark Pixel Percentages
        total_pixels = pixel_data.size
        for threshold in thresholds:
            dark_pixels = np.sum(pixel_data < threshold)
            metrics[f'img_Dark_Pixel_Percentage_less_than_{threshold}'] = (dark_pixels / total_pixels) * 100

        # 2. Brightness and Contrast
        metrics['img_Mean_Brightness'] = np.mean(pixel_data)
        metrics['img_Contrast'] = np.std(pixel_data)

        # 3. Texture Analysis
        metrics['img_Variance'] = np.var(pixel_data)
        metrics['img_Entropy'] = entropy(np.histogram(pixel_data, bins=256)[0])
        metrics['img_Energy'] = np.sum(pixel_data ** 2) / total_pixels
        # Calculating Homogeneity (using Co-occurrence Matrix)
        glcm = feature.greycomatrix(pixel_data, distances=[1], angles=[0], symmetric=True, normed=True)
        metrics['img_Homogeneity'] = feature.greycoprops(glcm, 'homogeneity')[0, 0]

        # 4. Edge Detection Metrics
        edges = filters.sobel(pixel_data)
        metrics['img_Edge_Density'] = np.mean(edges > edges.mean())
        metrics['img_Mean_Gradient_Magnitude'] = np.mean(np.abs(edges))
        
        # 5. Sharpness (Laplacian Variance)
        laplacian = filters.laplace(pixel_data)
        metrics['img_Sharpness'] = np.var(laplacian)

        # 6. Histogram-Based Metrics
        histogram, _ = np.histogram(pixel_data, bins=256, range=(0, 256))
        metrics['img_Histogram_Peaks'] = np.count_nonzero(histogram > histogram.mean())
        metrics['img_Histogram_Skewness'] = skew(pixel_data.flatten())
        metrics['img_Histogram_Kurtosis'] = kurtosis(pixel_data.flatten())
        
        # Colorfulness is not applicable for grayscale; you may skip this metric for grayscale images.

        return metrics
    except Exception as e:
        return {'Error': f"Error processing image: {e}"}

```

```python
# Apply the function to each row in the DataFrame and expand the result into separate columns
metrics_df = df['jp2'].parallel_apply(lambda x: calculate_image_features(x['bytes'])).apply(pd.Series)

# Combine the results with the original DataFrame
df = pd.concat([df, metrics_df], axis=1)

df
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Separating features and IDs for PCA processing
features = data.drop(columns=['well_id'])

# Standardizing the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Performing PCA on the scaled features
pca = PCA()
pca.fit(scaled_features)

# Transforming the data using PCA
pca_data = pca.transform(scaled_features)

# Creating a DataFrame for PCA results
pca_df = pd.DataFrame(pca_data, columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])
pca_df['well_id'] = data['well_id']

# Explained variance of each PCA component
explained_variance = pca.explained_variance_ratio_

# Displaying explained variance percentages
explained_variance_percentages = explained_variance * 100
explained_variance_percentages
```

```python
pca_df
```

```python
results = pd.merge(pca_df, df_all_channels, right_on="Converted_ID", left_on="well_id")
results
```

```python


# Calculate correlations with 'Dark_Pixel_Percentage_less_than_20'
correlations = finalset10.corr()['img_Histogram_Skewness_2']

# Display only feature correlations
correlations = correlations.drop('img_Histogram_Skewness_2')
correlations
```

```python
# Sort data for better visualization
sorted_correlation = dict(sorted(correlations.items(), key=lambda item: item[1]))

# Plotting
plt.figure(figsize=(6, 20))
plt.barh(list(sorted_correlation.keys()), list(sorted_correlation.values()))
plt.xlabel('Correlation with img_Histogram_Skewness_2')
plt.ylabel('Features')
plt.title('Correlation of Features with img_Histogram_Skewness_2')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Selecting dark pixel columns and feature columns for correlation analysis
dark_pixel_columns = [col for col in finalset10.columns if 'img_' in col]
feature_columns = [col for col in finalset10.columns if 'PC' in col]

# Calculating the correlation between dark pixel columns and feature columns
correlation_matrix = finalset10[dark_pixel_columns + feature_columns].corr()

# Extract correlations of dark pixel columns with individual features
correlation_with_dark_pixels = correlation_matrix.loc[dark_pixel_columns, feature_columns]
```

```python
correlation_with_dark_pixels = correlation_with_dark_pixels.iloc[:, :10]
correlation_with_dark_pixels
```

```python
import seaborn as sns


# Creating a clustered heatmap for the correlation matrix
plt.figure(figsize=(10, 50))
# Replacing any non-finite values with 0 to avoid errors in clustering
correlation_with_dark_pixels_cleaned = correlation_with_dark_pixels.replace([np.inf, -np.inf], np.nan).fillna(0)

# Creating the clustered heatmap again with cleaned data
sns.clustermap(correlation_with_dark_pixels_cleaned, cmap='coolwarm', annot=False, cbar=True)
plt.title("Clustered Correlation Heatmap of Dark Pixels with Individual Features", pad=20)
plt.show()
```
