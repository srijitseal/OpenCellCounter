```python
import numpy as np
import matplotlib.pyplot as plt
from dadapy.metric_comparisons import MetricComparisons
from dadapy.plot import plot_inf_imb_plane

# Step 1: Generate two sample datasets with 300 features each
np.random.seed(42)
N = 100  # number of samples
data1 = np.random.rand(N, 300)  # First dataset
data2 = np.random.rand(N, 300)  # Second dataset

# Step 2: Initialize the MetricComparisons class for data1
maxk = N - 1
metric_comparison_data1 = MetricComparisons(data1, maxk=maxk)

# Define coordinate subsets for comparison
coord_list = [[i] for i in range(300)]  # Individual feature comparison
labels = [f"Feature_{i}" for i in range(300)]

# Step 3: Compute information imbalance for each feature in data2
imbalances = metric_comparison_data1.return_inf_imb_full_selected_coords(coord_list, k=1)

# Step 4: Plot the information imbalance plane
plot_inf_imb_plane(imbalances, coord_list, labels)
plt.show()

# Display imbalances for specific interpretation
print("Information Imbalances between data1 and each feature in data2:", imbalances)
```
